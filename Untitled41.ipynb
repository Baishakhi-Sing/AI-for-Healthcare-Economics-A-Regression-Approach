```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Generate sample healthcare dataset since we don't have the actual file
np.random.seed(42)  # For reproducibility

# Number of samples
n_samples = 1000

# Generate features
age = np.random.randint(18, 65, n_samples)
bmi = np.round(np.random.normal(26, 6, n_samples), 1)
children = np.random.randint(0, 6, n_samples)
sex = np.random.choice(['male', 'female'], n_samples)
smoker = np.random.choice(['yes', 'no'], n_samples, p=[0.2, 0.8])
region = np.random.choice(['northeast', 'northwest', 'southeast', 'southwest'], n_samples)

# Generate expenses based on features (realistic pattern)
expenses = (
    200 * age +
    300 * bmi +
    500 * children +
    (10000 if 'yes' else 0) * (smoker == 'yes') +
    np.random.normal(0, 1000, n_samples)  # Random noise
)

# Ensure expenses are positive
expenses = np.abs(expenses)

# Create DataFrame
dataset = pd.DataFrame({
    'age': age,
    'sex': sex,
    'bmi': bmi,
    'children': children,
    'smoker': smoker,
    'region': region,
    'expenses': expenses
})

print("Generated Dataset:")
print(dataset.head())
print(f"\nDataset shape: {dataset.shape}")

# Step 1: Explore the data
print("\n=== Data Exploration ===")
print("Dataset info:")
print(dataset.info())

print("\nStatistical summary:")
print(dataset.describe())

print("\nCategorical value counts:")
print("Sex distribution:")
print(dataset['sex'].value_counts())
print("\nSmoker distribution:")
print(dataset['smoker'].value_counts())
print("\nRegion distribution:")
print(dataset['region'].value_counts())

# Step 2: Prepare the data
print("\n=== Data Preparation ===")

# Separate features and target
X = dataset.drop('expenses', axis=1)
y = dataset['expenses']

# Identify categorical and numerical columns
categorical_cols = ['sex', 'smoker', 'region']
numerical_cols = ['age', 'bmi', 'children']

# Create preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(drop='first'), categorical_cols)
    ])

# Apply preprocessing
X_processed = preprocessor.fit_transform(X)

# Split the data (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.2, random_state=42
)

print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")

# Step 3: Build and train the model
print("\n=== Model Training ===")
model = LinearRegression()
model.fit(X_train, y_train)

# Check model coefficients
print("Model trained successfully!")
print(f"Number of coefficients: {len(model.coef_)}")

# Step 4: Evaluate the model
print("\n=== Model Evaluation ===")

# Training set evaluation
train_predictions = model.predict(X_train)
train_mae = mean_absolute_error(y_train, train_predictions)
train_mse = mean_squared_error(y_train, train_predictions)
train_r2 = r2_score(y_train, train_predictions)

print("Training Results:")
print(f"Mean Absolute Error: ${train_mae:.2f}")
print(f"Mean Squared Error: ${train_mse:.2f}")
print(f"R² Score: {train_r2:.4f}")

# Test set evaluation
test_predictions = model.predict(X_test)
test_mae = mean_absolute_error(y_test, test_predictions)
test_mse = mean_squared_error(y_test, test_predictions)
test_r2 = r2_score(y_test, test_predictions)

print("\nTest Results:")
print(f"Mean Absolute Error: ${test_mae:.2f}")
print(f"Mean Squared Error: ${test_mse:.2f}")
print(f"R² Score: {test_r2:.4f}")

# Check if MAE is under 3500 (requirement to pass)
if test_mae < 3500:
    print("\n✅ SUCCESS: Model meets the requirement (MAE < $3500)")
else:
    print("\n❌ FAIL: Model does not meet the requirement (MAE >= $3500)")

# Step 5: Visualize the results
print("\n=== Visualization ===")

plt.figure(figsize=(15, 10))

# Plot 1: Actual vs Predicted
plt.subplot(2, 2, 1)
plt.scatter(y_test, test_predictions, alpha=0.6, color='blue')
max_val = max(y_test.max(), test_predictions.max())
plt.plot([0, max_val], [0, max_val], 'r--', lw=2)
plt.xlabel('Actual Expenses ($)')
plt.ylabel('Predicted Expenses ($)')
plt.title('Actual vs Predicted Expenses')
plt.grid(True, alpha=0.3)

# Plot 2: Residual plot
plt.subplot(2, 2, 2)
residuals = y_test - test_predictions
plt.scatter(test_predictions, residuals, alpha=0.6, color='green')
plt.axhline(y=0, color='r', linestyle='--', lw=2)
plt.xlabel('Predicted Expenses ($)')
plt.ylabel('Residuals ($)')
plt.title('Residual Plot')
plt.grid(True, alpha=0.3)

# Plot 3: Error distribution
plt.subplot(2, 2, 3)
plt.hist(residuals, bins=30, edgecolor='black', alpha=0.7, color='orange')
plt.axvline(x=0, color='r', linestyle='--', lw=2)
plt.xlabel('Prediction Error ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Prediction Errors')
plt.grid(True, alpha=0.3)

# Plot 4: Feature importance (coefficients)
plt.subplot(2, 2, 4)
# Get feature names after one-hot encoding
feature_names = (numerical_cols +
                 list(preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)))
coefficients = pd.DataFrame({
    'feature': feature_names,
    'coefficient': model.coef_
}).sort_values('coefficient', key=abs, ascending=False)

plt.barh(coefficients['feature'], coefficients['coefficient'])
plt.xlabel('Coefficient Value')
plt.title('Feature Importance (Coefficients)')
plt.tight_layout()

plt.show()

# Step 6: Make predictions on new data
print("\n=== Example Predictions ===")

def predict_healthcare_cost(age, sex, bmi, children, smoker, region):
    """
    Predict healthcare costs for a new individual
    """
    # Create a DataFrame with the input data
    new_data = pd.DataFrame({
        'age': [age],
        'sex': [sex],
        'bmi': [bmi],
        'children': [children],
        'smoker': [smoker],
        'region': [region]
    })

    # Preprocess the new data
    new_data_processed = preprocessor.transform(new_data)

    # Make prediction
    prediction = model.predict(new_data_processed)

    return prediction[0]

# Example predictions
examples = [
    (30, 'female', 22.0, 0, 'no', 'southwest'),
    (45, 'male', 30.0, 2, 'yes', 'northeast'),
    (25, 'male', 24.0, 1, 'no', 'northwest'),
    (60, 'female', 28.0, 3, 'no', 'southeast')
]

for i, (age, sex, bmi, children, smoker, region) in enumerate(examples, 1):
    prediction = predict_healthcare_cost(age, sex, bmi, children, smoker, region)
    print(f"Example {i}: {age}y/o {sex}, BMI: {bmi}, {children} children, "
          f"smoker: {smoker}, region: {region}")
    print(f"   Predicted cost: ${prediction:,.2f}\n")

print("=== Model Summary ===")
print(f"Final MAE: ${test_mae:.2f}")
print(f"Model performance: {'PASS' if test_mae < 3500 else 'FAIL'}")
print(f"R² Score: {test_r2:.4f} (closer to 1 is better)")
```
